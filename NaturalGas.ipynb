```python
def graph_NG(list):
    global clean_df
    clean_df=[]
    list_names=["Residential-NG","Commercial-NG", "Industrial-NG"]
    w=0
    for i in list:
        i = i.drop(["Unnamed: 1","map","linechart","units","source key"], axis=1)
        i=i.iloc [1:]
        list_regions=i["remove"].to_list()
        i=i.T
        i.columns = i.iloc[0] 
        i = i.drop("remove")
        clean_df.append(i)
        plt.figure(figsize=(16, 8), dpi=150)
        for j in list_regions:
            style.use('fivethirtyeight')
            graph=i[j].plot(label=j)
            graph.tick_params(axis = 'both', which = 'major', labelsize = 18)
            # adding title to the plot
        plt.title(list_names[w])
        plt.ylabel('Dollar per Cubic Foot')

        w=w+1
  
        # adding Label to the x-axis
        plt.xlabel('Date')
  
        # adding legend to the curve
        plt.legend()
```

```python
graph_NG(NG_Price_list)
```

```python
list_names=["Residential-Total NG Consumption","Commercial-Total NG Consumption", "Industrial-Total NG Consumption"]
w=0
for i in clean_df:
    i=i.sum(axis=0)
    i.plot.bar(rot=30, title=list_names[w])
    plt.show(block=True)
    w=w+1
```
