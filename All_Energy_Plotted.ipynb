```python
def graph_Egen(list):
    global clean_df
    w=0
    list_names=["United States","New England","New York","Mid-Atlantic","Southeast","Florida","Mid-continetial",
            "Southwest Power Pool","Texas","Northwest & Rockies","Southwest","California"]
    clean_df=[]
    for i in list:
        i = i.drop(["Unnamed: 1","map","linechart","units","source key"], axis=1)
        i=i.iloc [0:-2]
        list_regions=i["remove"].to_list()
        i=i.T
        i.columns = i.iloc[0] 
        i = i.drop("remove")
        clean_df.append(i)
        plt.figure(figsize=(16, 8), dpi=150)
        for j in list_regions:
            i[j].plot(label=j)
            # adding title to the plot
        plt.title(list_names[w])
        w=w+1
  
        # adding Label to the x-axis
        plt.xlabel('Date')
  
        # adding legend to the curve
        plt.legend()
```

```python
graph_Egen(list_regions_egen)
```
